Usare TensorFlow per costruire una NN (per breast cancer classification).

In the first cell we will upload files that contain the pre-sorted normalized data (from Winsconsin breast cancer dataset, sorted into training and test sets with the "input" values to the NN as "X" values, and the expected "output" (a 0 if benign and a 1 if malignant) as the Y values.
(lanciato il code, posso fare "sfoglia" e caricare i file csv)

```
import pandas as pd
from google.colab import files
file = files.upload()
X_train = pd.read_csv("xtrain.csv", header=None)
Y_train = pd.read_csv("ytrain.csv", header=None)
X_test = pd.read_csv("xtest.csv", header=None)
Y_test = pd.read_csv("ytest.csv", header=None)
```

Now I'm going to use Keras to define a NN that will be trained off of this data. This NN can then be used to predict future values for breast cancer presence.

```
from keras.models import Sequential
from keras.layers import Dense

classifier = Sequential() #Initialising the ANN

classifier.add(Dense(units = 16, activation = 'relu', input_dim = 30))
classifier.add(Dense(units = 8, activation = 'relu'))
classifier.add(Dense(units = 6, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))
```

Once the NN is defined, I now have to specify is the optimizer and loss function

```
classifier.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy')

classifier.summary() 
print("Done")
```

You now train the NN using Classifier.fit, passing it the training data -- i.e. for this set of X, this is what the Y should look like. The NN will then spot the patterns in the data, and build a neural network that could replciate that.

```
classifier.fit(X_train, Y_train, batch_size = 1, epochs = 100)
```

To predict new values, the NN uses classifier.predict. I'm going to pass it the test values for X (which the NN hasn't previously seen) and it will give me back a set of predicitions. These predicitions will be probabilities, so I will clean them up by saying that if they are greater than .5, I'mm make them 1 else I'll make them 0.

```
Y_pred = classifier.predict(X_test)
Y_pred = [ 1 if y >= 0.5 else 0 for y in Y_pred ]
print("Done")
```

Now we can loop through the set of predictions for the test set (called Y_pred) and the actual values for the test set (called Y_test), and see how alike they are -- if they are the same, I'll increment 'correct', otherwise I'll increment 'wrong'.
You'll see that result is 100% accurate, even though the NN reported a lower accuracy than that. Why? (NB la NN copiata da me ha una precisione migliore sul set di training)

```
total = 0
correct = 0
wrong = 0
#for i in Y_pred:
for i in range(len(Y_pred)):
  total += 1
  if(Y_test.at[i, 0] == Y_pred[i]):
    correct += 1
  else:
    wrong += 1
  
print("Total " + str(total))
print("Correct " + str(correct))
print("Wrong " + str(wrong))
```
