#Test perdita di precisione

Testo se a seguito della divisione con 65535.0, fare il cast a float32 perde di precisione.

```
%tensorflow_version 1.x
```

```
import numpy as np
from numpy import random
# Online Python compiler (interpreter) to run Python online.
# Write Python 3 code in this online editor and run it.

print("Hello world")


x=random.randint(65535, size=(10))
y = x
z = y
k = z

print(x)
print(y)
print(z)
print(k)

print("====")

x = x / 65535.0
print("Float")
print(x)

y = y / 65535.0
y = y.reshape(-1).astype(np.single)
print("Single")
print(y)

z = z / 65535.0
z = z.reshape(-1).astype(np.double)
print("Double")
print(z)


np.longdouble
k = k / 65535.0
k = k.reshape(-1).astype(np.longdouble)
print("Long double")
print(k)
```
