```
import glob
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import correlate2d
from skimage.util import compare_images
```

```
def comp_img(img1, img2):
  diff_rotated = compare_images(img1, img2, method='diff')
  plt.figure(figsize=(10,10))
  plt.imshow(diff_rotated)
  plt.grid(False)
  plt.title("Diff method")
  plt.show()


def correlate(img1, img2):
  diff_rotated = correlate2d(img1, img2)
  plt.figure(figsize=(10,10))
  plt.imshow(diff_rotated)
  plt.grid(False)
  plt.title("Cross correlation")
  plt.show()

#euclidean distance
def similar_perc(img1, img2):
  picture1 = img1
  picture2 = img2
  picture1_norm = picture1/np.sqrt(np.sum(picture1**2))
  picture2_norm = picture2/np.sqrt(np.sum(picture2**2))
  val = np.sum(picture2_norm*picture1_norm)
  val = round(val, 6)
  val = val*100
  print(str(val)+"%")

def confronti(img1, img2):
  comp_img(img1, img2)

  correlate(img1, img2)

  similar_perc(img1, img2)
```
